```python
# For Google Colab only: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Set the path to the image in your Google Drive
# Example path - adjust to your actual image location in Drive
img_path = '/content/drive/My Drive/me.jpg'

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# Load the image
img = mpimg.imread(img_path)

# Function to display 30 images in a 5x6 grid
def display_grid(image, rows=5, cols=6):
    fig, axes = plt.subplots(rows, cols, figsize=(12, 8))
    for i in range(rows):
        for j in range(cols):
            axes[i, j].imshow(image)
            axes[i, j].axis('off')
    plt.tight_layout()
    plt.show()

# Function to display images in a cross pattern
def display_cross(image):
    # Create an empty 5x5 grid for cross
    cross_grid = np.empty((5, 5), dtype=object)

    # Set the middle row and middle column to be the image
    for i in range(5):
        cross_grid[i, 2] = image  # Vertical part of the cross
        cross_grid[2, i] = image  # Horizontal part of the cross

    # Display the cross pattern
    fig, axes = plt.subplots(5, 5, figsize=(10, 10))
    for i in range(5):
        for j in range(5):
            if cross_grid[i, j] is not None:
                axes[i, j].imshow(cross_grid[i, j])
            axes[i, j].axis('off')
    plt.tight_layout()
    plt.show()

# Display the 5x6 grid
display_grid(img)

# Display the cross pattern
display_cross(img)
```
