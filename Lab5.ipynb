```python
import pandas as pd
import numpy as np
from scipy.stats import gmean, hmean, mode, skew
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
# Replace 'file_path.csv' with your file's path
data = pd.read_csv("sample_data.csv")

# Display first few rows
print("Dataset preview:")
print(data.head())

# Select numerical columns for analysis
numerical_data = data.select_dtypes(include=[np.number])
print("\nNumerical Columns:")
print(numerical_data.columns)

# Function to calculate and print descriptive statistics
def print_descriptive_stats(data):
    print("\n--- Descriptive Statistics ---")
    print(data.describe())
# Correlation Matrix
print("\n--- Correlation Matrix ---")
correlation_matrix = numerical_data.corr()
print(correlation_matrix)

# Plotting
print("\n--- Visualizations ---")

# Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=numerical_data)
plt.title("Boxplot of Numerical Features")
plt.show()

# Histograms
numerical_data.hist(bins=20, figsize=(15, 10), color="blue", alpha=0.7)
plt.suptitle("Histograms of Numerical Features")
plt.show()

# Density Plots
numerical_data.plot(kind='density', subplots=True, layout=(3, 3), figsize=(15, 10), sharex=False)
plt.suptitle("Density Plots of Numerical Features")
plt.show()

# Scatterplot (between the first two numerical features)
if len(numerical_data.columns) > 1:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=numerical_data.columns[0], y=numerical_data.columns[1], data=numerical_data)
    plt.title(f"Scatterplot: {numerical_data.columns[0]} vs {numerical_data.columns[1]}")
    plt.xlabel(numerical_data.columns[0])
    plt.ylabel(numerical_data.columns[1])
    plt.show()

# Bar Chart (for a categorical column if present)
categorical_columns = data.select_dtypes(include=['object']).columns
if len(categorical_columns) > 0:
    for col in categorical_columns:
        plt.figure(figsize=(8, 6))
        data[col].value_counts().plot(kind='bar', color="green")
        plt.title(f"Bar Chart for {col}")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.show()
```
