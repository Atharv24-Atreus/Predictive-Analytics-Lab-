```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load the CSV file
df = pd.read_csv('sample.csv')

# 1. Identify missing values
print("Missing values:\n", df.isnull().sum())

# 2. Handle missing values
# For demonstration purposes, let's assume we introduce some missing values
df.loc[1, 'Age'] = None  # Introduce NaN in 'Age'
df.loc[3, 'Occupation'] = None  # Introduce NaN in 'Occupation'

# Fill missing values with appropriate strategies
df_filled = df.fillna({
    'Name': 'Unknown',
    'Age': df['Age'].mean(),  # Replace NaN with the mean age
    'Occupation': 'Unknown',
    'Salary': df['Salary'].median()  # Replace NaN with the median salary
})

print("\nDataFrame after filling missing values:\n", df_filled)

# Alternatively, drop missing values (uncomment if needed)
# df_dropped = df.dropna()
# print("\nDataFrame after dropping missing values:\n", df_dropped)

# Interpolate missing values (uncomment if needed)
# df_interpolated = df.interpolate()
# print("\nDataFrame after interpolation:\n", df_interpolated)

# 3. Apply scaling techniques

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_filled[['Age', 'Salary']] = scaler_minmax.fit_transform(df_filled[['Age', 'Salary']])
print("\nDataFrame after Min-Max scaling:\n", df_filled)

# Z-score Normalization
scaler_zscore = StandardScaler()
df_filled[['Age', 'Salary']] = scaler_zscore.fit_transform(df_filled[['Age', 'Salary']])
print("\nDataFrame after Z-score normalization:\n", df_filled)

# 4. Create dummy variables
df_dummies = pd.get_dummies(df_filled, columns=['Occupation'])
print("\nDataFrame with dummy variables:\n", df_dummies)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Atharv', 'Aryan', 'Pratham', 'Manav', 'Bhavya'],
    'Age': [30, 25, 40, 35, 29],
    'Occupation': ['Engineer', 'Data Scientist', 'Doctor', 'Teacher', 'Artist'],
    'Salary': [70000, 80000, 120000, 60000, 55000]
}
df = pd.DataFrame(data)

# Group by 'Occupation' and calculate summary statistics
grouped = df.groupby('Occupation').agg({
    'Age': ['mean', 'median', 'count'],
    'Salary': ['mean', 'median', 'sum']
})

print("\nSummary statistics by Occupation:\n", grouped)
# Create a pivot table
pivot_table = pd.pivot_table(
    df,
    values='Salary',
    index='Occupation',
    aggfunc={'Salary': ['mean', 'sum']}
)

print("\nPivot Table:\n", pivot_table)
# Sample DataFrames
df1 = pd.DataFrame({
    'Employee': ['Atharv', 'Aryan', 'Pratham'],
    'Age': [30, 25, 40]
})

df2 = pd.DataFrame({
    'Employee': ['Aryan', 'Pratham', 'Manav'],
    'Salary': [80000, 120000, 60000]
})

# Concatenate along rows
df_concat_rows = pd.concat([df1, df2], axis=0, ignore_index=True)
print("\nConcatenated DataFrame (rows):\n", df_concat_rows)
# Concatenate along columns (requires matching indices)
df_concat_columns = pd.concat([df1, df2], axis=1)
print("\nConcatenated DataFrame (columns):\n", df_concat_columns)

# Merge DataFrames on 'Employee' column
df_merge = pd.merge(df1, df2, on='Employee', how='inner')
print("\nMerged DataFrame (inner join):\n", df_merge)
# Set 'Employee' as index for join demonstration
df1.set_index('Employee', inplace=True)
df2.set_index('Employee', inplace=True)

# Join DataFrames on index
df_join = df1.join(df2, how='inner')
print("\nJoined DataFrame (inner join):\n", df_join)
```
